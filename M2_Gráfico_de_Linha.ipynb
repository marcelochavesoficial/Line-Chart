**Line Chart**

A line graph is a type of graph that shows information in the form of a set of points (markers) connected by straight line segments. It is one of the basic graph types most used in most problems.

```python
# Import packages:
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(42)
x = np.arange(10)
y = np.random.normal(size=10)

# Plot the values:
plt.plot(x, y)
plt.title("Example")
plt.xlabel("Axle x")
plt.ylabel("Axle y")

plt.show()
```

Line charts are ideal for showing data changes and trends over equal time intervals.

When you want to observe the behavior of data over time, you use the x axis (horizontal) to represent the chronological order and the y axis for the dependent variable.

**When to use?**

To record data that connects and has continuity.

**Tips**

* Gestalt connection principle

        "Gestalt says that elements that are visually connected are perceived as more related than unconnected elements. This means that 2 identical figures in a layout, when disconnected, are seen differently than their misshapen figures connected by a line. Of all the principles that suggest that objects are related, the uniform connection tends to be the strongest."

* Doesn't make sense for categorical variables
* Often the X axis is a unit of time.

**Plot the line chart**

To exemplify the line graph, I will use a dataset containing the number of cases (monthly) of dengue for the city of Campinas-SP.

In addition to the number of confirmed dengue cases, the dataset also shows data related to rainfall, average, minimum and maximum temperatures in the municipality. The period covered by the data set is from 1998 to 2015.

```python
# Import data to dataframe
data = 'https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/dengue-dataset.csv'
df = pd.read_csv(data, index_col="data", parse_dates=True)

# See the first 5 entries
df.head()
```

```python
# Plot the number of confirmed cases
fig, ax = plt.subplots()

df["casos-confirmados"].plot(kind="line", ax=ax)
ax.set_title("Dengue em Campinas")
ax.set_ylabel("Casos Confirmados")
ax.set_xlabel("Data")

plt.tight_layout()
```

**Select ranges**

```python
# Plot the average temperature histogram
fig, ax = plt.subplots()

df.loc['2011-01-01':'2013-12-31',"temperatura-media"].plot(kind="line", ax=ax)
ax.set_title("Average Temperature Graph")
ax.set_ylabel("Confirmed Cases")
ax.set_xlabel("Date")

plt.tight_layout
```

**Plot multiple graphs separately**

```python
# Plot the average temperature histogram
# fig, ax = plt.subplots(nrows=2, nools=1, figsize=(8,4)) # ax[0] e ax[1]
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10,4))

# Plot average temperature
df.loc['2010-01-01':'2013-12-31',"temperatura-media"].plot(kind="line", ax=ax1)
ax2.set_title("Average Temperature Graph")
ax2.set_ylabel("Confirmed Cases")
ax2.set_xlabel("Date")

plt.tight_layout()
```

**Plot multiple overlapping graphs**

```python
# Plot the normalized graphs
df_norm = (df - df.mean()) / df.std()
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,4))

# Plot average temperature
df_norm.loc['2010-01-01':'2013-12-31',"temperatura-media"].plot(kind="line", ax=ax)

# Plot confirmed cases
df_norm.loc['2010-01-01':'2013-12-31':,"casos-confirmados"].plot(kind="line", ax=ax)

# x-axis and title
ax.set_xlabel("Date")
ax.set_title("Average Temperature X Confirmed Cases")

plt.tight_layout()
```
